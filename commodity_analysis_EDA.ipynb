```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df=df.sort_values('Date').reset_index(drop=True)
```

```python
print(df.head())
```

```python
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month
```

```python
df.columns
```

```python
missing_values=df.isnull().sum()
miss_data=missing_values[missing_values>0].sort_values(ascending=False)
```

```python
print(miss_data)
```

```python
print(len(miss_data))
```

```python
df_ffill=df.ffill()
```

```python
miss_values=df_ffill.isnull().sum()
miss_data=miss_values[miss_values>0].sort_values(ascending=False)
print(len(miss_data))
```

```python
df_clean=df.ffill().bfill()
```

```python
miss_values=df_clean.isnull().sum()
miss_data=miss_values[miss_values>0].sort_values(ascending=False)
print(len(miss_data))
```
